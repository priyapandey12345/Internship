```python
#1. Write a python program which searches all the product under a particular product from www.amazon.in. The 
#product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for 
#guitars. 


from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException


def search_product(product):
    
    driver = webdriver.Chrome()
    
    # Navigate to the Amazon website
    driver.get("https://www.amazon.in/")
    
    try:
# Find the search input field and enter the product name
        search_input = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "twotabsearchtextbox")))
        search_input.send_keys(product)
        
        # Submit the search form
        search_input.submit()
        
        # Wait for the search results to load
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "s-result-list")))
        
        # Print the search results
        search_results = driver.find_elements(By.CSS_SELECTOR, ".s-result-item .a-link-normal")
        for result in search_results:
            print(result.text)
        
    except TimeoutException:
        print("Timeout occurred while searching for the product.")
        
    

# Get the product name from the user
product_name = input("Enter the product name: ")


search_product(product_name)
```

```python
#2.In the above question, now scrape the following details of each product listed in first 3 pages of your searchresults and save it in a data frame and csv. In case if any product has less than 3 pages in search results then 
scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 

import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException

def scrape_amazon_products(product_name):
    # Create a WebDriver instance (e.g., Chrome WebDriver)
    driver = webdriver.Chrome()

    try:
        # Initialize data lists
        brand_names = []
        product_names = []
        prices = []
        return_exchange = []
        expected_delivery = []
        availabilities = []
        product_urls = []

        # Loop through the first 3 pages of search results
        for page_number in range(1, 4):
            url = f"https://www.amazon.in/s?k={product_name.replace(' ', '+')}&page={page_number}"
            driver.get(url)

            # Wait for the page to load (you may need to adjust the wait time)
            driver.implicitly_wait(10)

            # Find product containers
            product_containers = driver.find_elements_by_css_selector("div.s-include-content-margin")

            for product in product_containers:
                try:
                    brand = product.find_element_by_css_selector("span.a-size-base-plus.a-color-base.a-text-normal")
                    brand_names.append(brand.text.strip())
                except NoSuchElementException:
                    brand_names.append("-")

                try:
                    product_name = product.find_element_by_css_selector("span.a-text-normal")
                    product_names.append(product_name.text.strip())
                except NoSuchElementException:
                    product_names.append("-")

                try:
                    price = product.find_element_by_css_selector("span.a-offscreen")
                    prices.append(price.text.strip())
                except NoSuchElementException:
                    prices.append("-")

                try:
                    return_policy = product.find_element_by_css_selector("span.a-declarative")
                    return_exchange.append(return_policy.text.strip())
                except NoSuchElementException:
                    return_exchange.append("-")

                try:
                    delivery = product.find_element_by_css_selector("span.a-text-bold")
                    expected_delivery.append(delivery.text.strip())
                except NoSuchElementException:
                    expected_delivery.append("-")

                try:
                    availability = product.find_element_by_css_selector("span.a-size-base")
                    availabilities.append(availability.text.strip())
                except NoSuchElementException:
                    availabilities.append("-")

                try:
                    product_url = product.find_element_by_css_selector("a.a-link-normal.s-no-outline")
                    product_urls.append(product_url.get_attribute("href"))
                except NoSuchElementException:
                    product_urls.append("-")

        # Create a DataFrame
        data = {
            "Brand Name": brand_names,
            "Name of the Product": product_names,
            "Price": prices,
            "Return/Exchange": return_exchange,
            "Expected Delivery": expected_delivery,
            "Availability": availabilities,
            "Product URL": product_urls,
        }

        df = pd.DataFrame(data)

        # Save to CSV
        df.to_csv("amazon_products.csv", index=False)

        print("Data scraped and saved to 'amazon_products.csv'")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

if __name__ == "__main__":
    product_name = input("Enter the product you want to search for on Amazon.in: ")
    scrape_amazon_products(product_name)
```

```python
#3.Write a python program to access the search bar and search button on images.google.com and scrape 10 
images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’. 

import time
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains

# Function to scrape images for a given keyword
def scrape_images_for_keyword(keyword, num_images=10):
    # Create a WebDriver instance (e.g., Chrome WebDriver)
    driver = webdriver.Chrome()

    try:
        # Navigate to Google Images
        driver.get("https://www.google.com/imghp")

        # Find the search bar element and enter the keyword
        search_box = driver.find_element(By.NAME, "q")
        search_box.send_keys(keyword)

        # Submit the search
        search_box.send_keys(Keys.RETURN)

        # Scroll down to load more images (you may need to adjust the scroll count)
        for _ in range(3):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)

        # Find image elements
        image_elements = driver.find_elements(By.XPATH, "//img[@class='rg_i Q4LuWd']")

        # Create a directory to save the images
        if not os.path.exists(keyword):
            os.makedirs(keyword)

        # Download the images
        for i, img_element in enumerate(image_elements[:num_images], start=1):
            img_url = img_element.get_attribute("src")
            if img_url and not img_url.startswith("data:image"):
                img_extension = img_url.split(".")[-1]
                img_filename = f"{keyword}/{keyword}_{i}.{img_extension}"

                # Right-click on the image to open the context menu and select "Save image as..."
                ActionChains(driver).context_click(img_element).send_keys(Keys.ARROW_DOWN).send_keys(Keys.RETURN).perform()
                time.sleep(2)

                # Rename the downloaded file
                os.rename("image.png", img_filename)

                print(f"Downloaded image {i}/{num_images} for '{keyword}'")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

if __name__ == "__main__":
    keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']
    for keyword in keywords:
        scrape_images_for_keyword(keyword, num_images=10)
```

```python
#4.Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com
and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand 
Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, 
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
details is missing then replace it by “- “. Save your results in a dataframe and CSV

import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

def scrape_flipkart_smartphones(keyword):
    # Create a WebDriver instance (e.g., Chrome WebDriver)
    driver = webdriver.Chrome()

    try:
        # Navigate to Flipkart
        driver.get("https://www.flipkart.com/")

        # Close the login popup if it appears
        try:
            driver.find_element(By.CLASS_NAME, "_2KpZ6l._2doB4z").click()
        except Exception:
            pass

        # Find the search bar element and enter the keyword
        search_box = driver.find_element(By.NAME, "q")
        search_box.send_keys(keyword)
        search_box.send_keys(Keys.RETURN)

        # Wait for the page to load (you may need to adjust the wait time)
        driver.implicitly_wait(10)

        # Initialize data lists
        brand_names = []
        smartphone_names = []
        colors = []
        ram_list = []
        storage_list = []
        primary_cameras = []
        secondary_cameras = []
        display_sizes = []
        battery_capacities = []
        prices = []
        product_urls = []

        # Find product containers
        product_containers = driver.find_elements(By.CLASS_NAME, "_4ddWXP")

        for product in product_containers:
            # Extract details for each product
            brand_name = product.find_element(By.CLASS_NAME, "_4rR01T").text
            smartphone_name = product.find_element(By.CLASS_NAME, "IRpwTa").text
            details = product.find_elements(By.CLASS_NAME, "IRpwTa")
            color = details[0].text if len(details) > 0 else "-"
            ram = details[1].text if len(details) > 1 else "-"
            storage = details[2].text if len(details) > 2 else "-"
            primary_camera = details[3].text if len(details) > 3 else "-"
            secondary_camera = details[4].text if len(details) > 4 else "-"
            display_size = details[5].text if len(details) > 5 else "-"
            battery_capacity = details[6].text if len(details) > 6 else "-"
            price = product.find_element(By.CLASS_NAME, "_30jeq3").text
            product_url = product.find_element(By.TAG_NAME, "a").get_attribute("href")

            # Append details to the lists
            brand_names.append(brand_name)
            smartphone_names.append(smartphone_name)
            colors.append(color)
            ram_list.append(ram)
            storage_list.append(storage)
            primary_cameras.append(primary_camera)
            secondary_cameras.append(secondary_camera)
            display_sizes.append(display_size)
            battery_capacities.append(battery_capacity)
            prices.append(price)
            product_urls.append(product_url)

        # Create a DataFrame
        data = {
            "Brand Name": brand_names,
            "Smartphone Name": smartphone_names,
            "Colour": colors,
            "RAM": ram_list,
            "Storage(ROM)": storage_list,
            "Primary Camera": primary_cameras,
            "Secondary Camera": secondary_cameras,
            "Display Size": display_sizes,
            "Battery Capacity": battery_capacities,
            "Price": prices,
            "Product URL": product_urls,
        }

        df = pd.DataFrame(data)

        # Save to CSV
        df.to_csv("flipkart_smartphones.csv", index=False)

        print("Data scraped and saved to 'flipkart_smartphones.csv'")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

    finally:
        # Close the WebDriver
        driver.quit()

if __name__ == "__main__":
    keyword = input("Enter the smartphone you want to search for on Flipkart: ")
    scrape_flipkart_smartphones(keyword)
```

```python
#5.Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

def get_coordinates(city_name):
    # Create a WebDriver instance (e.g., Chrome WebDriver)
    driver = webdriver.Chrome()

    try:
        # Open Google Maps
        driver.get("https://www.google.com/maps")

        # Find the search input field
        search_box = driver.find_element_by_id("searchboxinput")

        # Enter the city name and submit the search
        search_box.send_keys(city_name)
        search_box.send_keys(Keys.RETURN)

        # Wait for the coordinates to load (you may need to adjust the wait time)
        time.sleep(5)

        # Find the coordinates in the URL
        url = driver.current_url
        lat_long_start = url.find('@') + 1
        lat_long_end = url.find(',', lat_long_start)
        latitude = url[lat_long_start:lat_long_end]
        longitude = url[lat_long_end + 1: url.find(',', lat_long_end + 1)]

        print(f"Coordinates for '{city_name}':")
        print(f"Latitude: {latitude}")
        print(f"Longitude: {longitude}")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

    finally:
        # Close the WebDriver
        driver.quit()

if __name__ == "__main__":
    city_name = input("Enter the city name to find its coordinates on Google Maps: ")
    get_coordinates(city_name)
```
